```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as clr
```

```python
def sign():
    if np.random.rand() < 0.5:
        return -1
    return 1
```

```python
def generate_points(N):
    Xs=[]
    Ys=[]
    i = 0
    while i < N:
        x = np.random.rand()*sign()
        y = np.random.rand()*sign()
        if x**2 + y **2 < 1:
            i += 1
            Xs.append(x)
            Ys.append(y)
    return (np.array(Xs), np.array(Ys))
```

```python
def RGB_to_hex(RGB):
    RGB = [int(x) for x in RGB]
    return "#"+"".join(["0{0:x}".format(v) if v < 16 else "{0:x}".format(v) for v in RGB])
```

```python
def graph_d2(x, y, fx, fy):
    N = len(x)
    col = [RGB_to_hex([255, int(255*(x[i]**2+y[i]**2)), 0]) for i in range(N)]
    plt.scatter(fx, fy, c=col)
    plt.show()
```

```python
def f(x0, y0):
    return (x0/(x0**2-1), y0/(y0**2-1))
```

```python
N=800
x, y = generate_points(N)
graph_d2(x, y, x, y)
```

```python
fy = [f(x[i], y[i])[1] for i in range(N)]
fx = [f(x[i], y[i])[0] for i in range(N)]
graph_d2(x, y, fx, fy)
```
